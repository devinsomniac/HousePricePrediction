```python
import numpy as np
rooms_ix,bedrooms_ix,population_ix,household_ix = 3,4,5,6
class CombinedAttributeAdder():
    def __init__(self):
        pass
    def fit(self,X,Y=None):
        return self
    def transform(self,X,Y=None):
        rooms_per_households = X[:,rooms_ix]/X[:,household_ix]
        population_per_household = X[:,population_ix]/X[:,household_ix]
        bedrooms_per_household = X[:,bedrooms_ix]/X[:,household_ix]
        bedroom_to_room = X[:,bedrooms_ix]/X[:,rooms_ix]
        return np.c_[X,rooms_per_households,population_per_household,bedrooms_per_household,bedroom_to_room]
```

```python
OriginalData['income_cat'] = pd.cut(OriginalData['median_income'],bins=[0,1.5,3,4.5,6,np.inf],labels=[1,2,3,4,5])
```

```python
X = OriginalData.drop(columns=['median_house_value'],axis=1)
Y = OriginalData['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42,stratify=OriginalData['income_cat'])
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler

numeric_features = ['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income']
categoric_feature = ['ocean_proximity']

Base_numeric_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('addColumns',CombinedAttributeAdder()),
])

Base_categoric_transformer = Pipeline(steps=[
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])


base_preprocessor = ColumnTransformer(
    transformers=[
        ('num',Base_numeric_transformer,numeric_features),
        ('cat',Base_categoric_transformer,categoric_feature)
    ]
)

X_train_Base = base_preprocessor.fit_transform(X_train)
X_test_Base = base_preprocessor.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
linear_pipeline = Pipeline(steps=[
    ('scaler',StandardScaler()),
    ('model',LinearRegression())
])

full_poly_pipeline = Pipeline(steps=[
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

decesion_tree_pipeline = Pipeline(steps=[
    ('scaler',StandardScaler()),
    ('model',DecisionTreeRegressor())
])

random_forest_pipeline = Pipeline(steps=[
    ('scaler',StandardScaler()),
    ('model',RandomForestRegressor())
])
```

```python
#Using traditional Machine Learning Models
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linear_pipeline.fit(X_train_Base,Y_train)
Y_pred_linear = linear_pipeline.predict(X_test_Base)
rmse_linear = np.sqrt(mean_squared_error(Y_test, Y_pred_linear))
mae_linear = mean_absolute_error(Y_test, Y_pred_linear)
r2_linear = r2_score(Y_test, Y_pred_linear)
print("Linear Regression RMSE:", rmse_linear)
print("Linear Regression MAE:", mae_linear)
print("Linear Regression R^2:", r2_linear)
```

```python
full_poly_pipeline.fit(X_train_Base,Y_train)
Y_poly_pred = full_poly_pipeline.predict(X_test_Base)
rmse_poly = np.sqrt(mean_squared_error(Y_test,Y_poly_pred))
mae_poly = mean_absolute_error(Y_test,Y_poly_pred)
r2_poly = r2_score(Y_test,Y_poly_pred)
print("Polynomial Regression RMSE:",rmse_poly)
print("Polynomial Regression MAE:",mae_poly)
print("Polynomial Regression R^2:",r2_poly)
```

```python
#Using MAachine Learning Models
decesion_tree_pipeline.fit(X_train_Base,Y_train)
Y_decision_tree_pred = decesion_tree_pipeline.predict(X_test_Base)
rmse_decision_tree = np.sqrt(mean_squared_error(Y_test,Y_decision_tree_pred))
mae_decision_tree = mean_absolute_error(Y_test,Y_decision_tree_pred)
r2_decision_tree = r2_score(Y_test,Y_decision_tree_pred)
print("Decision Tree RMSE:",rmse_decision_tree)
print("Decision Tree MAE:",mae_decision_tree)
print("Decision Tree R^2:",r2_decision_tree)
```

```python
random_forest_pipeline.fit(X_train_Base,Y_train)
Y_random_forest_pred = random_forest_pipeline.predict(X_test_Base)
rmse_random_forest = np.sqrt(mean_squared_error(Y_test,Y_random_forest_pred))
mae_random_forest = mean_absolute_error(Y_test,Y_random_forest_pred)
r2_random_forest = r2_score(Y_test,Y_random_forest_pred)
print("Random Forest RMSE:",rmse_random_forest)
print("Random Forest MAE:",mae_random_forest)
print("Random Forest R^2:",r2_random_forest)
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Predicted vs Actual
plt.scatter(Y_test, Y_random_forest_pred, color='blue', alpha=0.5, s=10, label='Data Points')

# Add a diagonal line (perfect prediction line)
min_val = min(min(Y_test), min(Y_random_forest_pred))
max_val = max(max(Y_test), max(Y_random_forest_pred))
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='Perfect Prediction')

# Labels and title
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs Actual Values (Random Forest)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
rmse_scorer = make_scorer(rmse, greater_is_better=False)
mae_scorer = make_scorer(mean_absolute_error,greater_is_better=False)
r2_scorer = make_scorer(r2_score)

pipelines = {
    'linear': linear_pipeline,
    'polynomial': full_poly_pipeline,
    'decision_tree': decesion_tree_pipeline,
    'random_forest': random_forest_pipeline
}

for name,pipeline in pipelines.items():
    print(f"\nEvaluating {name} with 5-fold Cross Validation (80% Training Points)")

    rmse_scores = -cross_val_score(pipeline,X_train_Base,Y_train,cv=5,scoring=rmse_scorer)
    print("Cross-Validation RMSE Scores:", rmse_scores)
    print("Average RMSE:", rmse_scores.mean())
    print("Standard Deviation of RMSE:", rmse_scores.std())
    print("\n")

    mae_scores = -cross_val_score(pipeline,X_train_Base,Y_train,cv=5,scoring=mae_scorer)
    print("Cross-Validation MAE Scores:", mae_scores)
    print("Average MAE:", mae_scores.mean())
    print("Standard Deviation of MAE:", mae_scores.std())
    print("\n")

    r2_scores = cross_val_score(pipeline,X_train_Base,Y_train,cv=5,scoring=r2_scorer)
    print("Cross-Validation R^2 Scores:", r2_scores)
    print("Average R^2:", r2_scores.mean())
    print("Standard Deviation of R^2:", r2_scores.std())
    print("\n")
```

```python
for name, pipeline in pipelines.items():
    print(f"\nFinal Evaluation of {name} on Test Set:")
    pipeline.fit(X_train_Base, Y_train)
    Y_pred = pipeline.predict(X_test_Base)
    rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
    mae = mean_absolute_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)
    print("Test RMSE:", rmse)
    print("Test MAE:", mae)
    print("Test R:", r2)
```

```python
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV
param_dist = {
    'model__n_estimators': randint(50, 200), 
    'model__max_depth': [None, 10, 20, 30],  
    'model__min_samples_split': [2, 5, 10],   
    'model__min_samples_leaf': [1, 2, 4]      
}

random_Search = RandomizedSearchCV(
    estimator=random_forest_pipeline,
    param_distributions=param_dist,
    n_iter=20,  
    cv=5,  
    scoring='neg_mean_absolute_error',  
    n_jobs=-1,  
    random_state=42
)

random_Search.fit(X_train_Base, Y_train)

# Best parameters and score from Randomized Search
print("Best Parameters:", random_Search.best_params_)
print("Best Cross-Validation MAE (from Randomized Search):", -random_Search.best_score_)
```

```python
best_rf_model = random_Search.best_estimator_
Y_rf_tuned_pred = best_rf_model.predict(X_test_Base)
rmse_rf_tuned = np.sqrt(mean_squared_error(Y_test, Y_rf_tuned_pred))
mae_rf_tuned = mean_absolute_error(Y_test, Y_rf_tuned_pred)
r2_rf_tuned = r2_score(Y_test, Y_rf_tuned_pred)
print("\nTuned Random Forest on Test Set:")
print("Test RMSE:", rmse_rf_tuned)
print("Test MAE:", mae_rf_tuned)
print("Test R:", r2_rf_tuned)
```
